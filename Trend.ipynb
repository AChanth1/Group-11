```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/Users/abigaelawino/Downloads/datasetforModel (1).xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Extract the year from the Execution Date
df['Execution Year'] = df['Execution Date'].dt.year

# Group by Execution Year and sum the Execution Volunteer counts
trend_data = df.groupby('Execution Year')['Execution Volunteer'].sum().reset_index()

# Plot the trend line
plt.figure(figsize=(12, 8))
sns.lineplot(data=trend_data, x='Execution Year', y='Execution Volunteer', marker='o', color='blue')
plt.title("Trend of Execution Volunteers Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Execution Volunteers", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
